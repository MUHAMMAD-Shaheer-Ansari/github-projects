```python
model = YOLO('yolov8n.pt')

cap = cv2.VideoCapture('/content/pexels-paul-daley-3318088-1920x1080-50fps.mp4')
success = True
imgs=[]

while success:
  success ,img = cap.read()
  if success == False:
    break
  result = model(img,stream = True)
  for r in result:
    boxes = r.boxes
    for box in boxes:
      x1,y1,x2,y2 = box.xyxy[0]
      x1,y1,x2,y2= int(x1),int(y1),int(x2),int(y2)
      if int(box.cls[0]) == 0:
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,75,0),7)
        cv2.putText(img,'person',(max(0,x1),max(35,y1-20)),cv2.FONT_HERSHEY_COMPLEX,2,(0,0,0),7)
  imgs.append(img)
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
size = list(imgs[0].shape)
size = size[:2]
to_append = size[0]
size.remove(size[0])
size.append(to_append)
video = cv2.VideoWriter('/content/test.mp4',fourcc,30,size)

for i in range(len(imgs)):
  video.write(imgs[i])

video.release()
```
